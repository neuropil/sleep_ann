```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pt1_df = pd.read_pickle('pt1_results.df')
pt2_df = pd.read_pickle('pt2_results.df')
merge_df = pd.read_pickle('ts_merge_results.df')
```

```python
pt1_df['set']=['pt1']*len(pt1_df)
pt2_df['set']=['pt2']*len(pt2_df)
merge_df['set']=['combined']*len(merge_df)
all_df = pd.concat([pt1_df,pt2_df,merge_df]).reset_index()
sns.set_context('talk')
sns.set(font_scale=2)
sns.set_style(style='whitegrid')
ax = sns.stripplot(x='set',y='accuracy',palette='plasma',data=all_df,jitter=True)
ax.set_ylim(0.5,1)
sns.despine(ax=ax)
```

```python
sns.set(font_scale=2)
sns.set_style(style='whitegrid')
ax = sns.violinplot(x='set',y='accuracy',palette='plasma',data=all_df,jitter=True)
ax.set_ylim(0.5,1)
sns.despine(ax=ax)
```
